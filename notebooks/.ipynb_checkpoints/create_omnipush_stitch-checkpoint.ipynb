```python
%matplotlib inline
import matplotlib
import numpy as np
import matplotlib.pyplot as plt
import glob
import os
from scipy.misc import imread
from skimage.transform import resize
import imageio
from IPython.display import Image
from IPython.display import display
from IPython.display import clear_output
from tqdm import tqdm
```

## Function for visualization

```python
def inspect_seqs(seq_dirs):
    ims = []
    for seq_dir in seq_dirs:
        fnames = glob.glob(os.path.join(seq_dir, '*.png'))
        for i in range(len(fnames)):
            # image
            fname = os.path.join(seq_dir, '{}.png'.format(i))
            im = imageio.imread(fname).reshape(64, 64, 3)
            ims.append(im)

    clear_output(wait=True)
    if os.path.exists('./tmp.gif'):
        os.remove('./tmp.gif')
    imageio.mimsave('./tmp.gif', ims)
    with open('./tmp.gif','rb') as f:
        display(Image(data=f.read(), format='gif', width=200, height=200))
```

## Define AlexNet for extracting feature for stitching video

```python
import torch
import torch.nn as nn
import torchvision.models as models

alexnet = models.alexnet(pretrained=True)

class AlexNetConv4(nn.Module):
    def __init__(self):
        super(AlexNetConv4, self).__init__()
        self.features = nn.Sequential(
            # stop at conv4
            *list(alexnet.features.children())[:-3]
        )
    def forward(self, x):
        x = self.features(x)
        return x

alexnet_conv4 = AlexNetConv4()
alexnet_conv4.cuda()
alexnet_conv4.eval()
```

```python
def preprocess(imgs):
    imgs_new = []
    for img in imgs:
        img_tmp = img.astype(np.float32) / 255.0
        img_tmp = (img_tmp - [0.485, 0.456, 0.406]) / [0.229, 0.224, 0.225]
        imgs_new.append(resize(img_tmp, (224, 224)))
    imgs_prcoessed = torch.from_numpy(np.array(imgs_new)).view(-1, 3, 224, 224).float().cuda()
    return imgs_prcoessed
```

## Get stitching info

```python
def get_seqs_shape(DATASET_PATH, split):
    seqs_shape = {}
    seqs_more_than_two = []
    seqs_l2_loss_weird = []
    shape_dirs = sorted(glob.glob(os.path.join(DATASET_PATH, '{}/**/'.format(split))))
    for shape_dir in tqdm(shape_dirs):
        shape_name = shape_dir.split('{}/'.format(split))[1].split('/motion')[0][:-1]
        seq_dirs = sorted(glob.glob(os.path.join(shape_dir, '*')))
        n_rep = int(seq_dirs[-1].split('rep=')[1].split('_push')[0]) + 1
        seqs_stitch = []
        for rep in range(n_rep):
            rep_dirs = sorted([rep_dir for rep_dir in seq_dirs if 'rep={:04d}'.format(rep) in rep_dir])
            n_push = int(rep_dirs[-1].split('push=')[1].split('_t=')[0]) + 1

            seq_stitch_1 = []
            seq_stitch_2 = []

            for push in range(n_push):
                push_dirs = [push_dir for push_dir in rep_dirs if 'rep={:04d}_push={:04d}'.format(rep, push) in push_dir]
                if len(push_dirs) > 2:
                    seqs_more_than_two.append(push_dirs)
                    break
                elif len(push_dirs) == 2:
                    if len(seq_stitch_1) == 0 and len(seq_stitch_2) == 0:  # start of an ambiguous seq
                        seq_stitch_1.append(push_dirs[0])
                        seq_stitch_2.append(push_dirs[1])
                    elif len(seq_stitch_1) != 0 and len(seq_stitch_2) == 0:
                        img_start_1 = imread(os.path.join(push_dirs[0], '0.png'))
                        img_start_2 = imread(os.path.join(push_dirs[1], '0.png'))
                        x = preprocess([img_end_1, img_start_1, img_start_2])
                        features = alexnet_conv4(x).detach().cpu().numpy()
                        e1 = features[0]
                        s1 = features[1]
                        s2 = features[2]

                        err_e1s1 = np.mean((e1 - s1) ** 2)
                        err_e1s2 = np.mean((e1 - s2) ** 2)

                        if err_e1s1 < err_e1s2:
                            seq_stitch_1.append(push_dirs[0])
                            seq_stitch_2.append(push_dirs[1])
                        else:
                            seq_stitch_1.append(push_dirs[1])
                            seq_stitch_2.append(push_dirs[0])
                    else:
                        img_start_1 = imread(os.path.join(push_dirs[0], '0.png'))
                        img_start_2 = imread(os.path.join(push_dirs[1], '0.png'))

                        x = preprocess([img_end_1, img_end_2, img_start_1, img_start_2])
                        features = alexnet_conv4(x).detach().cpu().numpy()
                        e1 = features[0]
                        e2 = features[1]
                        s1 = features[2]
                        s2 = features[3]

                        err_e1s1 = np.mean((e1 - s1) ** 2)
                        err_e1s2 = np.mean((e1 - s2) ** 2)
                        err_e2s1 = np.mean((e2 - s1) ** 2)
                        err_e2s2 = np.mean((e2 - s2) ** 2)

                        seq_next_idx_1 = np.argmin([err_e1s1, err_e1s2])
                        seq_next_idx_2 = np.argmin([err_e2s1, err_e2s2])

                        if seq_next_idx_1 != seq_next_idx_2:
                            seq_stitch_1.append(push_dirs[seq_next_idx_1])
                            seq_stitch_2.append(push_dirs[seq_next_idx_2])                        
                        else:  # handle cases when l2 error doesn't work
                            seqs_l2_loss_weird.append(push_dirs)
                            if seq_next_idx_1 == 0:
                                if err_e1s1 < err_e2s1:
                                    seq_stitch_1.append(push_dirs[0])
                                    seq_stitch_2.append(push_dirs[1])
                                else:
                                    seq_stitch_1.append(push_dirs[1])
                                    seq_stitch_2.append(push_dirs[0])
                            else:
                                if err_e1s2 < err_e2s2:
                                    seq_stitch_1.append(push_dirs[1])
                                    seq_stitch_2.append(push_dirs[0])
                                else:
                                    seq_stitch_1.append(push_dirs[0])
                                    seq_stitch_2.append(push_dirs[1])

                    img_end_1 = imread(os.path.join(seq_stitch_1[-1], '11.png'))
                    img_end_2 = imread(os.path.join(seq_stitch_2[-1], '11.png'))
                elif len(push_dirs) == 1:
                    if len(seq_stitch_2) == 0:
                        seq_stitch_1.append(push_dirs[0])
                        img_end_1 = imread(os.path.join(seq_stitch_1[-1], '11.png'))
                    else:
                        img_start = imread(os.path.join(push_dirs[0], '0.png'))
                        err_e1s = np.mean((img_end_1 - img_start) ** 2)
                        err_e2s = np.mean((img_end_2 - img_start) ** 2)
                        if err_e1s < err_e2s:
                            seq_stitch_1.append(push_dirs[0])
                            img_end_1 = imread(os.path.join(seq_stitch_1[-1], '11.png'))
                        else:
                            seq_stitch_2.append(push_dirs[0])
                            img_end_2 = imread(os.path.join(seq_stitch_2[-1], '11.png'))

            if len(seq_stitch_1) > 0:
                seqs_stitch.append(seq_stitch_1)
            if len(seq_stitch_2) > 0:
                seqs_stitch.append(seq_stitch_2)

        seqs_shape[shape_name] = seqs_stitch
    return seqs_shape

DATASET_PATH = '/data/vision/phillipi/gen-models/svg/dataset/omnipush'
seqs_shape = get_seqs_shape(DATASET_PATH, 'train')
```

## Save all the stitch seqs


```python
def save_seqs(seq_dirs):
    shape_name = seq_dirs[0].split('train/')[1].split('/motion')[0]
    d = './stitch_videos/{}'.format(shape_name)
    f = seq_dirs[0].split('/')[-1]
    
    if not os.path.exists(d):
        os.makedirs(d)
    
    ims = []
    for seq_dir in seq_dirs:
        fnames = glob.glob(os.path.join(seq_dir, '*.png'))
        for i in range(len(fnames)):
            # image
            fname = os.path.join(seq_dir, '{}.png'.format(i))
            im = imageio.imread(fname).reshape(64, 64, 3)
            ims.append(im)

    imageio.mimsave('./stitch_videos/{}/{}.gif'.format(shape_name, f), ims)
```

```python
shape_names = os.listdir(os.path.join(DATASET_PATH, 'train/'))
for shape_name in tqdm(shape_names):
    for i in range(len(seqs_shape[shape_name])):
        save_seqs(seqs_shape[shape_name][i])
```

## Finally, it's time to creat a stitched dataset

```python
from shutil import copyfile
import numpy as np
split = 'train'
SRC_DATASET_NAME = 'omnipush'
TAR_DATASET_NAME = 'omnipush_stitch'
DATASET_PATH = os.path.join('/data/vision/phillipi/gen-models/svg/dataset/', SRC_DATASET_NAME)
OUTPUT_PATH = os.path.join('/data/vision/phillipi/gen-models/svg/dataset/', TAR_DATASET_NAME)
seqs_stitch = get_seqs_shape(DATASET_PATH, split)

shape_names = os.listdir(os.path.join(DATASET_PATH, split))
for shape_name in tqdm(shape_names):
    for seq_stitch in seqs_stitch[shape_name]:
        d = os.path.join(OUTPUT_PATH, seq_stitch[0])
        tar_d = d.replace(SRC_DATASET_NAME, TAR_DATASET_NAME)
        if not os.path.exists(tar_d):
            os.makedirs(tar_d)

        action_stitch = []
        for i, _ in enumerate(seq_stitch):
            action_src = np.load(os.path.join(seq_stitch[i], 'actions.npy'))
            action_stitch.append(action_src)
            for j in range(len(img_fnames)):
                img_src = os.path.join(seq_stitch[i], '{}.png'.format(j))
                img_tar = os.path.join(tar_d, '{}.png'.format(i * len(img_fnames) + j))
                copyfile(img_src, img_tar)
        np.save(os.path.join(tar_d, 'actions.npy'), np.array(action_stitch).reshape((-1, 3)))
```
